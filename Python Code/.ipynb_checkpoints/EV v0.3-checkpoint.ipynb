```python
import os
import cv2
import numpy as np
import random
import re  # Import the regular expressions module
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Activation
import matplotlib.pyplot as plt

import numpy as np
import tensorflow as tf

# Set a random seed (for reproducibility, set it to a specific value if needed)
#random_seed = random.randint(1, 1000)

# Set a fixed seed (constant seed)
random_seed = 2  # Replace with your desired seed value

# Set the random seed for NumPy
np.random.seed(random_seed)

# Set the random seed for TensorFlow
tf.random.set_seed(random_seed)


# Now you can use the `random_seed` variable to easily toggle between random and fixed seeds

# Define the path to your training data folder
data_folder = 'Ground Truthing Training Images'

# Initialize empty lists to store images and masks
images = []
masks = []

# Define a regular expression pattern to extract the numerical part of the filenames
pattern = r'\d+'  # Matches one or more digits

# Create a function to extract the numerical part from a filename
def extract_number(filename):
    return int(re.findall(pattern, filename)[0])  # Extract the first match and convert to int

# Get a list of all image and mask filenames
image_filenames = [filename for filename in os.listdir(data_folder) if filename.startswith('image_')]
mask_filenames = [filename for filename in os.listdir(data_folder) if filename.startswith('mask_')]

# Sort the filenames based on the numerical part (e.g., image_002.png will come before image_004.png)
image_filenames.sort(key=extract_number)
mask_filenames.sort(key=extract_number)

# Load the images and masks in the sorted order
for image_filename, mask_filename in zip(image_filenames, mask_filenames):
    # Load the image
    image_path = os.path.join(data_folder, image_filename)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Normalize the image to the range [0, 255]
    image = (image / 232.0) * 255.0  # Adjust the denominator as needed
    
    images.append(image)

    # Load the corresponding mask
    mask_path = os.path.join(data_folder, mask_filename)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    masks.append(mask)

# Convert the lists to numpy arrays
images = np.array(images)
masks = np.array(masks)

# Define the input shape as (512, 512, 1)
input_shape = (512, 512, 1)

# Split data
split = int(0.8 * len(images))
train_images, val_images = images[:split], images[split:]
train_masks, val_masks = masks[:split], masks[split:]

# Define the U-Net architecture with 512x512 output
def build_unet(input_shape):
    inputs = Input(input_shape)

    # Encoder (contracting path)
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    # Add more convolutional and pooling layers here...

    # Decoder (expansive path)
    up6 = UpSampling2D(size=(2, 2))(pool1)  # Upsample to match size

    # Add more upsampling and convolutional layers...

   # Output layer with 512x512 output shape without sigmoid activation
    out = Conv2D(1, 1, padding='same')(up6)


    model = Model(inputs=inputs, outputs=out)
    return model

# Create the U-Net model
model = build_unet(input_shape)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_images, train_masks, validation_data=(val_images, val_masks), epochs=10, batch_size=32)

# Visualize the predicted masks
num_samples = val_images.shape[0]
predicted_masks = model.predict(val_images)

# Threshold the predicted masks to make them binary
threshold = 0.5  # You can adjust this threshold as needed
binary_masks = (predicted_masks > threshold).astype(np.uint8) * 255

# Display original images, ground truth masks, and predicted masks
for i in range(num_samples):
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 3, 1)
    plt.imshow(val_images[i, :, :], cmap='gray')
    plt.title("Original Image")
    
    plt.subplot(1, 3, 2)
    plt.imshow(val_masks[i, :, :], cmap='gray')
    plt.title("Ground Truth Mask")
    
    plt.subplot(1, 3, 3)
    plt.imshow(binary_masks[i, :, :], cmap='gray')
    plt.title("Predicted Mask")
    
    plt.show()
```
