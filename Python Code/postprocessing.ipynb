# Highlighting points of interest

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
# load both the original and mask
mask = cv2.imread('../Example Images/ground_truth/gt_C-8.png')
original = cv2.imread('../Example Images/C-8.png')
```

```python
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
```

```python
highlighted = original + 1000
highlighted[mask == 0] = original[mask == 0]
highlighted[highlighted > 255] = 255
```

```python
plt.imshow(cv2.cvtColor(highlighted, cv2.COLOR_BGR2RGB))
```

```python
root = '../Example Images/'
orig_files = [x for x in os.listdir(root) if '.png' in x]
gt_files = [x for x in os.listdir(root + 'ground_truth')]
for i in range(len(orig_files)):
    orig_img = cv2.imread(root + orig_files[i])
    gt_img = cv2.imread(root + 'ground_truth/' + gt_files[i])
    
    highlighted = orig_img + 1000
    highlighted[gt_img == 0] = orig_img[gt_img == 0]
    highlighted[highlighted > 255] = 255
    
    cv2.imwrite(root + 'overlapped/ol_' + orig_files[i], highlighted.astype(np.uint8))
```
